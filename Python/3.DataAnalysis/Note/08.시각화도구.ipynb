# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['전입지별'].isnull()
```

```python
# NaN 전(Prev) 데이터로 채우기
    # 전출지별 전국으로 NaN값 채우기
df = df.fillna(method='ffill')
df.head()
```

```python
df.info()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
    # 전출지별이 서울특별시이고 전입지별이 서울특별시가 아닌곳
df_seoul = df[(df.전출지별 =='서울특별시')&(df.전입지별 !='서울특별시')]
df_seoul.head()
```

```python
len(df_seoul)
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop("전출지별", axis='columns', inplace=True)
```

```python
df_seoul.head()
```

```python
# 전입지별 -> '전입지' 로 컬럼 이름 변경
df_seoul.rename(columns={"전입지별": "전입지"}, inplace=True)
df_seoul.head()
```

```python
# df_seoul의 index를 전입지로 변경
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

```python
plt.plot(sr_one)
plt.show()
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트 제목 추가
plt.plot(sr_one)
plt.title("서울 -> 경기 인구 이동")

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 그래프 꾸미기
# 그림 사이즈 지정하기(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation='vertical')

# y축 눈금 라벨 회전하기
plt.yticks(rotation=45)

# 그래프 제목
plt.title('서울 -> 경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.plot(sr_one)

# 범례 표시
plt.legend(labels = ['서울 -> 경기'], loc = 'best')
plt.show()
```

```python
# 부산 -> 서울 인구이동 그래프 그리기

# 전출지:부산광역시, 전입지:서울특별시
df_BusantoSeoul = df[(df.전출지별 =='부산광역시')&(df.전입지별 =='서울특별시')]
# 전입지별 컬럼 이름을 전입지로 변경
df_BusantoSeoul.rename(columns={"전입지별": "전입지"}, inplace=True)
# 전입지 컬럼을 index로 설정
df_BusantoSeoul.set_index('전입지', inplace=True)
# 전출지별 컬럼 삭제
df_BusantoSeoul.drop("전출지별", axis='columns', inplace=True)
# 값만 가져오기
sr_two = df_BusantoSeoul.loc['서울특별시']
sr_two.head()
```

### 그래프에 Style 적용하기

```python
# 그래프 꾸미기
# 그림 사이즈 지정하기(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# Style 서식 지정
plt.style.use('ggplot')

# x축 눈금 라벨 회전하기
plt.xticks(rotation=45)

# y축 눈금 라벨 회전하기
plt.yticks(rotation=45)

# 그래프 제목
plt.title('부산 -> 서울 인구이동', size =20)
plt.xlabel('기간', size =15)
plt.ylabel('이동 인구수', size =15)

# Marker표시
plt.plot(sr_two, marker='o', markersize=10)

#   y축 범위 지정
plt.ylim(10000,80000) # ylim(최소값, 최대값)

# 화살표 표시
plt.annotate('',
            xy = (12,70000),   # 화살표의 머리부분
            xytext=(0, 35000),   # 화살표의 꼬리부분
            # xycoords='data',
            arrowprops=dict(arrowstyle='->', color= 'skyblue', lw=5)
            )
plt.annotate("인구 이동 증가(1970~1983)",
            xy = (5, 45000),
            rotation = 37,
            va = 'baseline', # 텍스트 상하정렬
            ha = 'center', # 텍스트 좌우정렬
            fontsize = 12
            )

plt.annotate('',
            xy = (48,30000),   # 화살표의 머리부분
            xytext=(15, 60000),   # 화살표의 꼬리부분
            # xycoords='data',
            arrowprops=dict(arrowstyle='->', color= 'olive', lw=5)
            )
plt.annotate("인구 이동 감소(1983~2017)",
            xy = (32, 45000),
            rotation = -15,
            va = 'baseline', # 텍스트 상하정렬
            ha = 'center', # 텍스트 좌우정렬
            fontsize = 12
            )


# 범례 표시
plt.legend(labels = ['부산 -> 서울'], loc = 'best', fontsize=15)
plt.show()

# 이미지 저장
plt.savefig('../Data/seoul.png')
```

### 동일한 그림에 여러개의 그래프를 추가


```python
# 서울에서 '충청남도', '강원도', '경상북도'로 이동한 인구데이터 값만 선택
df_3 = df_seoul.loc[['충청남도','경상북도','강원도']]
df_3
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

# 충청남도 그래프
ax.plot(df_3.columns,
        df_3.loc['충청남도'],
        marker='o',
        markerfacecolor= 'green',
        markersize= 10,
        color = 'olive',
        linewidth= 2,
        label= '서울->충남'
        )
# 경상북도 그래프
ax.plot(df_3.columns,
        df_3.loc['경상북도'],
        marker='o',
        markerfacecolor= 'blue',
        markersize= 10,
        color = 'skyblue',
        linewidth= 2,
        label= '서울->경북'
        )
# 강원도 그래프
ax.plot(df_3.columns,
        df_3.loc['강원도'],
        marker='o',
        markerfacecolor= 'red',
        markersize= 10,
        color = 'magenta',
        linewidth= 2,
        label= '서울->강원'
        )
plt.ylim(10000,80000)
ax.legend(loc='best')
ax.set_title('서울->충남, 경북, 강원 인구 이동', size=20)

ax.set_xlabel('기간',size=12)
ax.set_ylabel('이동 인구수',size=12)

ax.set_xticklabels(df_3.columns, rotation=90)

# 축 눈금 라벨 크기
ax.tick_params(axis='x', labelsize= 5)
ax.tick_params(axis='y', labelsize= 5)

plt.show()
```

```python
# 서울에서 '충청남도', '강원도', '경상북도', '전라남도' 로 이동한 인구데이터 값만 선택
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_4
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,2,3)
ax4 = fig.add_subplot(4,2,4)

# 충청남도 그래프
ax1.plot(df_4.columns,
        df_4.loc['충청남도'],
        marker='o',
        markerfacecolor= 'green',
        markersize= 10,
        color = 'olive',
        linewidth= 2,
        label= '서울->충남'
        )
# 경상북도 그래프
ax2.plot(df_4.columns,
        df_4.loc['경상북도'],
        marker='o',
        markerfacecolor= 'blue',
        markersize= 10,
        color = 'skyblue',
        linewidth= 2,
        label= '서울->경북'
        )
# 강원도 그래프
ax3.plot(df_4.columns,
        df_4.loc['강원도'],
        marker='o',
        markerfacecolor= 'red',
        markersize= 10,
        color = 'magenta',
        linewidth= 2,
        label= '서울->강원'
        )
# 전라남도 그래프
ax4.plot(df_4.columns,
        df_4.loc['전라남도'],
        marker='o',
        markerfacecolor= 'red',
        markersize= 10,
        color = 'magenta',
        linewidth= 2,
        label= '서울->전남'
        )

ax.legend(loc='best')
ax.set_title('서울->충남, 경북, 강원 인구 이동', size=20)

ax.set_xlabel('기간',size=12)
ax.set_ylabel('이동 인구수',size=12)

ax.set_xticklabels(df_4.columns, rotation=90)

# 축 눈금 라벨 크기
ax.tick_params(axis='x', labelsize= 5)
ax.tick_params(axis='y', labelsize= 5)

plt.show()
```
