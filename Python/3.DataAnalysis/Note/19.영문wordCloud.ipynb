```python
import nltk
import pickle
from nltk.corpus import stopwords
import re
nltk.download('all')
```

## text를 불러서 word 분류

```python
content = open("../Data/Independence.txt").read()
content
```

```python
# 불필요한 심벌 없애기
#cleaned_content = content.replace('!','').replace(',','')
cleaned_content = re.sub(r'[^\.\?\!\w\d\s]','',content) # 문장단위로 끊기
cleaned_content
```

```python
# 대문자를 소문자로 변환
cleaned_content = cleaned_content.lower()
cleaned_content
```

```python
# 문장에서 워드를 토큰으로 쪼갠다.
word_tokens = nltk.word_tokenize(cleaned_content)
print(word_tokens)
```

```python
# 품사 분리
token_pos = nltk.pos_tag(word_tokens)
print(token_pos)
```

```python
# 명사만 추출하기
NN_words = []
for word, pos in token_pos:
    # pos에 NN이 있으면
    if 'NN' in pos:
        NN_words.append(word)

print(NN_words)
```

```python
# 원형으로 복원
wlem = nltk.WordNetLemmatizer()
lenmatized_words = []
for word in NN_words:
    # 복수형(s)를 없애주기
    new_word = wlem.lemmatize(word)
    lenmatized_words.append(new_word)

print(lenmatized_words)
```

```python
# 불용어 제거
stopwords_list = stopwords.words('english')
# 중복값 제거 => set()
unique_NN_words = set(lenmatized_words)
final_NN_words = lenmatized_words

for word in unique_NN_words:
    if word in stopwords_list:
        while word in final_NN_words:
            final_NN_words.remove(word)
            
print(final_NN_words)
```

```python
# 원하는 단어 제거
customized_stopwords = ['be','today','yesterday',"it's", "don't",'others']

unique_NN_words1 = set(final_NN_words)
for word in unique_NN_words1:
    if word in customized_stopwords:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 빈도분석
    # RNN에서 많이 사용함
from collections import Counter

c = Counter(final_NN_words)
print(c)
```

```python
print(c.most_common(20))
```

```python
# 워드 클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wordcloud

noun_text = ''
for word in final_NN_words:
    noun_text = noun_text + ' ' + word

noun_text
```

```python
# 인스턴스 만들기
    # wordcloud 는 3차원 그래프임 가장 많이 나온 글자는 앞면에 적게나오면 뒷면에 보이면서 크기차이가 생겨보임
wordcloud = WordCloud(
    max_font_size=60,
    relative_scaling=0.5
).generate(noun_text) # generate가 noun_text의 string value를 입력받아 count함.

plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
