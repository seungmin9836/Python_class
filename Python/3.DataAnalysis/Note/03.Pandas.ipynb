### 행과 열을 이용한 데이터 선택

```python
exam_data = {'수학':[90,80,70],'영어':[98,89,95],
             '음악':[85,95,100], '체육':[100,90,90],
             '이름':['서준', '우현', '인아']}
df = pd.DataFrame(exam_data)
df
```

```python
# index이름을 정하기
df.set_index('이름', inplace=True)
df
```

```python
# loc #1
print(df.loc['서준',['음악','체육']])
# loc #2
print(df.loc['서준','음악':'체육'])
# loc #3
print(df.loc['서준','음악':])

# iloc #1
print(df.iloc[0,[2,3]])
# iloc #2  Slicing 응용
print(df.iloc[0,2:4])
# iloc #3
print(df.iloc[0,2:])
```

```python
# df에 컬럼(열)을 추가하기
df['국어'] = 80
df
```

```python
# df에 인덱스(행)을 추가하기
df.loc['철수'] = [10, 20, 30, 40, 50]
df
```

```python
# 영희를 추가할때 인아점수로 추가하기
df.loc['영희'] = df.loc['인아']
df
```

```python
# 서준의 체육점수를 80점으로 변경하기
    # loc
df.loc['서준','체육'] = 80
df
    # iloc
df.iloc[0,3] = 80
df
```

```python
# 서준의 음악과 체육점수를 50점으로 변경하기
    # loc
df.loc['서준',['체육','음악']] = 50
df
    # iloc
df.iloc[0,2:4] = 50
df
```

```python
# 서준의 음악 점수를 100점 체육점수를 50점으로 변경하기
    # loc 튜플활용 
df.loc['서준',['음악','체육']] = 100,50
df
    # iloc
df.iloc[0,2:4] = 100,50
df
```

```python
# 행과 열의 위치 바꾸기
    # transpose()
df.transpose()
```

```python
    # .T
df.T
```

```python
data = {'c0':[1,2,3],'c1':[4,5,6],
        'c2':[7,8,9], 'c3':[10,11,12],
        'c4':[13,14,15]
        }
df = pd.DataFrame(data, index=['r0','r1','r2'])
df
```

```python
# 인덱스를 재설정하기
new_index = ['r0','r1','r2','r3','r4']
ndf = df.reindex(new_index)
ndf
```

```python
# Nan값 0으로 대체하기 
ndf = df.reindex(new_index, fill_value=0)
ndf
```

```python
# reset_index()
    # 인덱스 초기화 함수
ndf = df.reset_index()
ndf
```

```python
df
```

```python
# 내림차순으로 행 인덱스 정렬
    # 기본값은 오름차순
df.sort_index(ascending=False)
```

```python
# 특정 값을 기준으로 정렬
df
```

```python
# c1 컬럼의 값들을 내림차순으로 정렬하기
    # 변경값 저장하기 : inplace = True
df.sort_values(by='c1', ascending=False, inplace=True)
```
