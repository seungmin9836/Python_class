```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
# Chrome Browser와 Chrome Driver Version 확인하기
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('https://web.dominos.co.kr/branch')
```

```python
# 자치구명 가져오기
# //*[@id="SIGUNGU_NM0"]
    # numeric이어서 ''홑따옴표로 써야되나봐
gu_list_raw = driver.find_element(By.XPATH, '//*[@id="SIGUNGU_NM0"]')
print(gu_list_raw)
gu_list = gu_list_raw.find_elements(By.TAG_NAME,'option')
# option 안에 있는 value들을 가져오기
gu_names = [option.get_attribute('value') for option in gu_list]
# list안에 있는 ' '빈칸 없애주기
gu_names.remove('')
gu_names
```

```python
# 시/도명 가져오기
cido_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_1"]')
print(cido_list_raw)
cido_list = cido_list_raw.find_elements(By.TAG_NAME,'option')
cido_names = [option.get_attribute('value') for option in cido_list]
print(cido_names)
```

```python
# 시/도명 가져오기
# //*[@id="region_code_1"]

cido_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_1"]')
cido_list = cido_list_raw.find_elements(By.TAG_NAME, 'option')  # option, value 이름등은 언제든지 바뀔 수 있음

cido_names = [option.accessible_name for option in cido_list]
cido_names.remove('시/도')
cido_names
```

```python
# 구/군명 가져오기
gugun_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_2"]')
print(gugun_list_raw)
gugun_list = gugun_list_raw.find_elements(By.TAG_NAME,'option')

gugun_names = [option.accessible_name for option in gugun_list]
gugun_names.remove('구/군')
gugun_names
```
