### BeautifulSoup 기본 사용법

```python
# Library

from bs4 import BeautifulSoup
import urllib.request as req

# Site Address

url = "http://zeushahn.github.io/Test/python/bs_exam01.html"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 원하는 부분 추출하기
    # beautifulsoup이 <태그> 기준으로 나눠서 가져와줌
h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling

print("h1 =",h1.string)
print("p1 =",p1.string)
print("p2 =",p2.string)
```

```python
# soup.find_all() : 다가져오기
links = soup.find_all("a")
#print(links)
    # list 구나 => for문을 돌리자
for a in links:
    # print(a)
    text = a.string
    href = a.attrs['href']
    print(text, ">", href)
```

### CSS 선택자 사용하기
- BeautifulSoup은 자바스크립트 라이브러리인 JQuery처럼 CSS선택자를 지정해서 원하는 요소를 추출 가능하다.

```python
# Library

from bs4 import BeautifulSoup
import urllib.request as req

# Site Address

url = "http://zeushahn.github.io/Test/python/bs_exam04.html"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 필요한 부분을 CSS 쿼리로 추출하기 (# : id를 의미함 / .(점) : class를 의미함 / > : 자식 / 빈칸 : 후손)
h1 = soup.select_one("div#meigen>h1").string
print(h1)
```

```python
li_list = soup.select("div#meigen > ul.items > li")
print(li_list)
```

```python
for li in li_list:
    print(li.string)
```

```python
# Library

from bs4 import BeautifulSoup
import urllib.request as req

# Site Address

url = "https://finance.naver.com/marketindex/"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 데이터 추출하기
price = soup.select_one("#exchangeList > li.on > a.head.usd > div > span.value").string
print(price)
```

# Exercise
: 미국, 일본, 유럽연합, 중국의 환율 가져오기


```python
# 데이터 추출하기
price_USA = soup.select_one("#exchangeList > li > a.head.usd > div > span.value").string
price_JAPAN = soup.select_one("#exchangeList > li > a.head.jpy > div > span.value").string
price_EU = soup.select_one("#exchangeList > li:nth-child(3) > a.head.eur > div > span.value").string
price_CHINA = soup.select_one("#exchangeList > li > a.head.cny > div > span.value").string
print("미국 \t: ",price_USA.rjust(10))
print("일본 \t: ",price_JAPAN.rjust(10))
print("유럽연합 \t: ",price_EU.rjust(10))
print("중국 \t: ",price_CHINA.rjust(10))
```

```python
price_list = soup.select("#exchangeList  span.value")
nations = soup.select("#exchangeList > li > a.head > h3 > span")

for i in range(len(price_list)):
    print('%s \t: %s'%(nations[i].string.ljust(10), price_list[i].string.rjust(10)))
```

```python
# Library

from bs4 import BeautifulSoup
import urllib.request as req

# Site Address

url = "https://movie.daum.net/ranking/boxoffice/yearly?date=2021"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
```

```python
movie_name = soup.select("strong.tit_item > a")
print(movie_name)
```

```python
movie_img = soup.select("div.poster_movie > img ")
for i in movie_img:
    href = i.attrs['src']
    print(href)
```

```python
for i in range(len(movie_name[:25])):  
    print(i+1,":",movie_name[i].string, movie_img[i].attrs['src'])
```

```python
# Exercise
# 파일 이름이 mov02.jpg, mov03.jpg, mov04.jpg
# 라이브러리 불러오기
import urllib.request as request

# 다운로드 파일 이름 결정
fileName = ['mov02.jpg', 'mov03.jpg', 'mov04.jpg']

# URL과 저장 경로 지정하기

for i in range(len(movie_name[:25])):  
    request.urlretrieve(movie_img[i].attrs['src'],f"../Data/{movie_name[i].string}.jpg")
print('ok')
```
