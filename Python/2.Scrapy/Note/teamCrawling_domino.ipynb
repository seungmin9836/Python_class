```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
import urllib.request as request
```

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('https://web.dominos.co.kr/branch')
```

```python
# 현재 화면 Capture 하기

driver.save_screenshot("../Data/Domino001.png")
```

```python
# 지역명 가져오기
# //*[@id="region_code_1"]

region_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_1"]')
region_list = region_list_raw.find_elements(By.TAG_NAME, 'option')  # option, value 이름등은 언제든지 바뀔 수 있음

region_names = [option.accessible_name for option in region_list]
region_names.remove('시/도')
region_names
```

```python
# 자치구명 가져오기
# //*[@id="region_code_2"]

gu_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_2"]')
gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')  # option, value 이름등은 언제든지 바뀔 수 있음

gu_names = [option.accessible_name for option in gu_list]
gu_names.remove('구/군')
gu_names
```

```python
# 매장조회버튼 click
# //*[@id="storeSrch1"]/div/div[1]/div[3]/button

xpath = '//*[@id="storeSrch1"]/div/div[1]/div[3]/button'
element_sel_region = driver.find_element(By.XPATH, xpath).click()
```

```python
# action selenium, 조회는 beautifulSoup
# 매장 찾기 버튼
# //*[@id="storeSrch1"]/div/div[1]/div[3]/button

# 지역 :  //*[@id="region_code_1"]
# 구군 :  //*[@id="region_code_2"]
import time
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
# 도미노피자 지점명
dominoMajangName = []

# 도미노피자 지점 주소
dominoAddress = []

# 지역별
for region in region_names:
    gu_names = []

    region_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_1"]')
    
    region_list_raw.send_keys(region)
    time.sleep(1)

    gu_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_2"]')
    gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')  # option, value 이름등은 언제든지 바뀔 수 있음

    gu_names = [option.accessible_name for option in gu_list]
    gu_names.remove('구/군')
    time.sleep(1)

    # 군/구 별
    for gu in gu_names:
        gu_list_raw = driver.find_element(By.XPATH, '//*[@id="region_code_2"]')
        gu_list_raw.send_keys(gu)
        
        # 매장찾기 버튼 클릭
        xpath = '//*[@id="storeSrch1"]/div/div[1]/div[3]/button'
        element_sel_gu = driver.find_element(By.XPATH,xpath).click()

        page_source = driver.page_source
        #  BeautifulSoup으로 분석하기
        soup = BeautifulSoup(page_source, "html.parser")

        # 매장/전화번호
        dominoName = soup.select("#ul_shop_list dt ")
        
        # 주소
        addressList = soup.select("#ul_shop_list dd.address")

        for i in range(len(addressList)):
            
            dominoMajangName.append(dominoName[i].text.split('점')[0].strip())
            dominoAddress.append(addressList[i].string.strip())


 
        
```
