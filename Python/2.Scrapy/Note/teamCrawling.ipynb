```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
import urllib.request as request
```

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('https://www.burgerking.co.kr/#/store')
```

```python
# 현재 화면 Capture 하기

driver.save_screenshot("../Data/Domino001.png")
```

```python
# 지역명 가져오기
# //*[@id="region_code_1"]
xpath = '//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[1]/ul/li[3]/button'
click_city = driver.find_element(By.XPATH, xpath).click()

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# 자치구명 옵션 리스트
city = soup.select("div.form_list select:nth-child(1) option")
city.pop(0)
city
city_s = []
for c in city:
    city_s.append(c.string.strip())
city_s
```

```python
# test
name = soup.select(".tit strong")
adds = soup.select(".addr span:nth-child(1)")

gameng = []
address = []

for i in range(len(name)):
    gameng.append(name[i].string.strip())
    address.append(adds[i].string.strip())

len(gameng)
```

```python
# 화면에 그림을 그리는 시간이 필요하므로 이름 delay 시키자!
import time
# 데이터프레임 뼈대 만들기
# df = pd.DataFrame(columns=['가맹점명','주소'])
gameng = []
address = []

for gu in city_s:
    # 시 선택
    xpath = '//*[@id="app"]/div/div[3]/div[3]/div/div[1]/div[1]/div[2]/div[4]/div/select[1]'
    element = driver.find_element(By.XPATH, xpath)
    element.send_keys(gu)

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    time.sleep(2)

    name = soup.select(".tit strong")
    adds = soup.select(".addr span:nth-child(1)")

    time.sleep(1)

    for i in range(len(name)):
            gameng.append(name[i].text.strip())
            address.append(adds[i].string.strip())
```

```python
# 저장하기
# 데이터 프레임 뼈대 만들기

df = pd.DataFrame(columns=['가맹점명','주소'])

for i in range(len(gameng)):
    df.loc[i] = [gameng[i], address[i]]

print(df)

 # 데이터 프레임 csv로 저장하기
df.to_csv('../Data/BurgerKing.csv', encoding='utf-8', index=False)
```
