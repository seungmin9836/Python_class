### Data 정규화

```python
# weight, height
csv['weight'] = csv['weight'] / csv['weight'].max()
csv['height'] = csv['height'] / csv['height'].max()
```

```python
csv.head()
```

```python
# label
    # one hot encoding
bclass = {'thin':[1,0,0], 'normal':[0,1,0], 'fat':[0,0,1]}
y = []
for i,v in enumerate(csv['label']):
    y.append(bclass[v])
y
```

```python
# Training과 Test
X_train = csv[['height', 'weight']][0:15000]
y_train = y[0:15000]
```

```python
# Test
X_test = csv[['height', 'weight']][15001:20001]
y_test = y[15001:20001]
```

## Deep Learning 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(512, activation ='relu', input_shape = (2,)))
# thin nomal fat => 3개
model.add(keras.layers.Dense(3, activation= 'softmax'))
```

```python
from pickletools import optimize
from sklearn import metrics

model.compile(
    # 'rmsprop', 'adam' 'sgd'
    optimizer = 'sgd',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy'],
)
```

```python
# 데이터 훈련하기
import numpy as np

model.fit(
    # 문제와 정답
    np.array(X_train),
    np.array(y_train),
    epochs = 50
)
```

```python
# 테스트 데이터로 평가하기
    # optimizer -> 3가지를 돌려보면서 loss율이 가장 적은 것이 과대적합이 일어날 가능성이 적다!
score = model.evaluate(X_test, np.array(y_test))
print('loss :', score[0])
print('accuracy :', score[1])
```
