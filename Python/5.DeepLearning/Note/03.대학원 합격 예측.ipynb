```python
# 결측치 데이터 삭제
data.dropna(inplace=True)

# 결측치데이터를 다른 값으로 대체
# data.fillna(100)
```

```python
# Training Data
x = data[['gre','gpa','rank']]
x.head()
```

```python
# Target Data
y = data['admit']
y.head()
```

```python
x.count()
```

### Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
# 1번째 Hidden LayerL 256노드 (2의 배수, 임의의 갯수 (Hyper Parameter))
# activation = sigmoid, tanh, relu, softmax, leakyRelu
model.add(keras.layers.Dense(256, activation = 'tanh', input_shape=(3,))) # 은닉층
model.add(keras.layers.Dense(1, activation='sigmoid')) # 출력층
```

```python
# 손실함수
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
# Optimaizere : learning rate를 알맞게조정해줌
# - adam
# - adagrad
# - adadelta
# - resprop
# - sgd
# - loss : 손실함수
# - binary_crossentropy : 분류 및 확률 예측
# - 
```

```python
# 학습시키기
import numpy as np

model.fit(np.array(x), np.array(y), epochs=500)
```

```python
# 예측하기

model.predict(
    [
        [700, 3.7, 1],
        [700, 3.7, 2],
        [700, 3.7, 3],
        [700, 3.7, 4],
    ]
)
```
